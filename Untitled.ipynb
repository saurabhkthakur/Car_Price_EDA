# UNIVARIATE ANALYSIS

```python

fig = plt.figure(figsize=(12,6))
plt.style.use('fast')

sns.barplot(x=df['make'].value_counts().values[:20],
           y=df['make'].value_counts().index[:20])

plt.tick_params(axis='x', labelsize=10)
plt.tick_params(axis='y', labelsize=15)
plt.xlabel('Car')
plt.ylabel('')
plt.title('Top 20 Cars present in data ')

plt.show()
```

```python
fig = plt.figure(figsize=(12,6))
plt.style.use('fast')

sns.barplot(x=df['segment'].value_counts().values,
           y=df['segment'].value_counts().index)

plt.tick_params(axis='x', labelsize=10)
plt.tick_params(axis='y', labelsize=15)
plt.xlabel('Car')
plt.ylabel('')
plt.title('Which kind of segment is most popular ')

plt.show()
```

```python
plt.rcParams['figure.figsize']= (15,7)

plt.subplot(1,2,1)
plt.style.use('fast')

sns.barplot(x=df['transmission'].value_counts().index,
           y=df['transmission'].value_counts(normalize=True).values)

plt.tick_params(axis='x', labelsize=10)
plt.tick_params(axis='y', labelsize=15)
plt.xlabel('Car')
plt.ylabel('')
plt.title('Popular Transmission Mode')


plt.subplot(1,2,2)
labels = df['drive_unit'].value_counts().index
size = df['drive_unit'].value_counts()
colors = ['lightgreen', 'orange', 'aqua', 'purple']
explode = [0.2, 0.3, 0.4,0.]


plt.pie(size, colors = colors, explode = explode, labels = labels, shadow = True, autopct = '%.2f%%')
plt.title('Drive_Unit', fontsize = 20)
plt.axis('off')
plt.legend()

plt.show()
```

```python
plt.rcParams['figure.figsize']= (15,7)

plt.subplot(1,2,1)
sns.set(style='whitegrid')

sns.distplot(df['priceUSD'], color='red')
plt.title('Price of Model in USD', fontsize=20)
plt.xlabel('range of USD')
plt.ylabel('count')


plt.subplot(1,2,2)
labels = df['condition'].value_counts().index
size = df['condition'].value_counts()
colors = ['lightgreen', 'orange', 'aqua']
explode = [0.2, 0.3, 1]


plt.pie(size, colors = colors, explode = explode, labels = labels, shadow = True, autopct = '%.2f%%')
plt.title('Condition of Car', fontsize = 20)
plt.axis('off')
plt.legend()
plt.show()
```
